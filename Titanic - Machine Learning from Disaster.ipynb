```python
import pandas as pd
```

```python
# creating a title column using their names 
title = []
for i in range(len(train_data["Name"])):
    if train_data["Sex"][i] == "male":
        title.append("Mr")
    else :
        if train_data["Age"][i] >= 30 :
            title.append("Mrs")
        else :
            title.append("Miss")
    
```

```python
train_data["title"] = title
```

```python
train_data = train_data.drop("Name",axis=1)
```

```python
train_data.head()
```

```python
train_data["Cabin"].fillna("Unknown",inplace=True)
```

```python
train_data
```

```python
cabin_list = ['A', 'B', 'C', 'D', 'E', 'F', 'T', 'G']
deck =[]

for i in range(len(train_data["Cabin"])):
    for j in cabin_list :
        if j in train_data["Cabin"][i] :
            deck.append(j)
            break
    
    else:
        deck.append("Unknown")
    
    
```

```python
d = pd.DataFrame(deck)
```

```python
d.value_counts()
```

```python
train_data["Cabin"].value_counts()
```

```python
train_data["Deck"] = deck
```

```python
train_data
```

```python
train_data = train_data.drop("Cabin",axis=1)
```

```python
train_data
```

```python
train_data.dtypes
```

```python
for label, content in train_data.items():
    if pd.api.types.is_string_dtype(content):
        train_data[label] = content.astype("category").cat.as_ordered()
```

```python
train_data.dtypes
```

```python
train_data.isna().sum()
```

```python
train_data["Embarked"].value_counts()
```

```python
train_data["Embarked"].fillna("S",inplace = True)
```

```python
train_data.isna().sum()
```

```python
train_data["Age"].median()
```

```python
train_data["Age"].fillna(28,inplace = True)
```

```python
train_data.isna().sum()
```

```python
train_data["Sex"].cat.codes
```

```python
# Turn categorical variables into numbers
for label, content in train_data.items():
    # Check columns which *aren't* numeric
    if not pd.api.types.is_numeric_dtype(content):
        # We add the +1 because pandas encodes missing categories as -1
        train_data[label] = pd.Categorical(content).codes+1
```

```python
train_data
```

```python
x = train_data.drop("Survived",axis=1)
y = train_data["Survived"]
```

```python
import numpy as np
np.random.seed(123)

from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(x,y)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
# testing in test data 

test_data = pd.read_csv("test.csv")
```
